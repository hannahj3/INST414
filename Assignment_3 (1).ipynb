```python
car_list = []
for i in range (2000, 2023):
    year = str(i)
    api_url = 'https://api.api-ninjas.com/v1/cars?year={}'.format(year)
    response = requests.get(api_url, headers={'X-Api-Key': 'xMZOW99mtQYVFe30+e3v8A==FEhFnY0MTnEUSZrG'})
    car_list.append(response.json())
    
```

```python
target_car = car_list[0][0]
```

```python
new_list = []
for year in car_list:
    for curr_car in year:
        ed_value = distance.euclidean([target_car['cylinders'],target_car['displacement'],target_car['highway_mpg']],
                           [curr_car['cylinders'],curr_car['displacement'],curr_car['highway_mpg']])
        curr_car['distance'] = ed_value
        new_list.append(curr_car)
        
```

```python
final_list = sorted(new_list, key=lambda x: x['distance']) 
```
